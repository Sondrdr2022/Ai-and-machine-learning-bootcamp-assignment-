```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import drive
drive.mount('/content/drive')

df = pd.read_excel("/content/drive/MyDrive/Bootcamp_project_Data.xlsx")

df
```

Function that take in the data file and print out the top 5 students with highest attendance.

```python
#Fillter out the column that start with "Week" in the dataframe
attendance= [col for col in df.columns if col.startswith('Week')]

# Adding all the values in each column to get the total attendance for each student
df['Total Attendance'] = df[attendance].sum(axis=1)

# Sort the students based on total attendance and select the top 5
top5= df[['First name', 'Total Attendance']].sort_values(by='Total Attendance', ascending=False).head()

print("Top 5 students with highest attendance:")
print(top5)
```

Function that take in the data file and print out the week with highest attendance.

```python
# Adding all the values in each column to get the total attendance per week
weekly_attendance = {week: df[week].sum() for week in attendance}

# Convert the dictionary to pandas data
total = pd.Series(weekly_attendance)

# Find the week with the highest attendance
highest_week = total.idxmax()
students = total.max()

print(f"The week with the highest attendance is {highest_week} with {students} students.")
```

Function that take in the data file and print out the number of faculties and the student per faculty

```python
#Finding the numbers of faculty
num_faculties = df['Faculty'].nunique()

#Count the number of students per faculty
students_per_faculty = df['Faculty'].value_counts()

print(f"Number of faculties: {num_faculties}")
print(f"Number of students per:{students_per_faculty}")
```

Function that take in the data file and print out the number of Students whose name Starts with M

```python
# Creates a boolean Series where True indicates the 'First name' starts with 'M'.
M =df['First name'].str.startswith('M').sum()

print(f"Number of students whose names start with M is: {M}")
```

Function that takes in the data frame as input and create an heat map of the entire data visualised.

```python
#Select data that is only a numerical number for the heat map
df_numeric = df.select_dtypes(include=['number'])

plt.figure(figsize=(10, 8))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm')
plt.title('Heat map of the entire data visualised.')
plt.xticks(rotation=50)
plt.show()
```
